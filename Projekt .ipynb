```python
ax=[]
Ax=[]
kx=[]
pp=[]
```

```python
def CzynnikDyskontujacy(i):
    return 1/(1+i)
```

```python
V=CzynnikDyskontujacy(0.1)
```

```python
def kPx(k,n,lx):
    p=[]
    for i in range(0,n-k):
        p.append(lx[30+k+i]/lx[30+k])
    return p
```

```python
def Ubez(k,n,px,qx):
    suma=0
    for i in range(0,n-k):
        suma +=V**(i+1)*px[i]*qx[30+i]
    return 10000*(suma + V**(n-k)*lx[30+(n-k)+k]/lx[30+k])
        
```

```python
def Renta(k,n,px):
    suma=0
    for i in range(0,n-k):
        suma += V**(i)*px[i]
    return suma
    
```

```python
def P(A,a):
    return(A/a)
```

```python
def rezerwa(A,P,a):
    return A-P*a
```

```python
p=kPx(0,20,lx)
A=Ubez(0,20,p,qx)
a=Renta(0,20,p)
```

```python
P=P(A,a)
```

```python
for k in range(0,20):
    p=kPx(k,20,lx)
    A=Ubez(k,20,p,qx)
    a=Renta(k,20,p)
    kv = rezerwa(A,P,a)
    ax.append(round(a,2))
    Ax.append(round(A,2))
    kx.append(round(kv,2))
```

```python
df = pd.DataFrame(zip(k,ax,Ax,kx),columns=['k','ax','Ax','kv'])
```
